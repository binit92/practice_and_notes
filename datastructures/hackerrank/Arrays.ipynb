### Dynamic Array
https://www.hackerrank.com/challenges/dynamic-array/problem

```python
arr = [int(arr_temp) for arr_temp in input().strip().split(' ')]

# the number of sequence 
N = arr[0]
# the number of queries
Q = arr[1]

# creating a list of N empty sequences
seqList = [[]] * N
lastAnswer = 0

for i in range(Q):
    arg =  input().split(' ')
    query = int(arg[0])
    x = int(arg[1])
    y = int(arg[2])
    
    if query == 1:
        index = (x^lastAnswer)%N
        seqList[index]= seqList[index] + [y]
    elif query == 2:
        index = (x^lastAnswer)%N
        seq = seqList[index]
        number = y%len(seq)
        lastAnswer = int(seqList[index][number])
        print(lastAnswer)
        
```

### Left Rotation
https://www.hackerrank.com/challenges/array-left-rotation/problem Given an array of  integers and a number, , perform  left rotations on the array. Then print the updated array as a single line of space-separated integers.


```python
def leftRotation(a, d):    
    b = [0] *len(a)    
    for x in range(len(a)):        
        if (x+d)<len(a):             
            b[x] = a[x+d]
        else:
            index = (x+d)%len(a)
            if not index>0:
                index = 0
            b[x] = a[index]                         
    return b    
       
if __name__ == "__main__":
    n, d = input().strip().split(' ')
    n, d = [int(n), int(d)]
    a = list(map(int, input().strip().split(' ')))
    result = leftRotation(a, d)
    print (" ".join(map(str, result)))
```

### Sparse Arrays
There is a collection of *N* strings ( There can be multiple occurences of a particular string ). Each string's length is no more than *20*  characters. There are also *Q* queries. For each query, you are given a string, and you need to find out how many times this string occurs in the given collection of *N* strings.
https://www.hackerrank.com/challenges/sparse-arrays/problem

```python
N = int(input())
strings = [None]*N
for i in range(N):
    strings[i] = input()

Q = int(input())
queries = [None]*Q
for j in range(Q):
    queries[j] = input()

for x in queries:
    count=0
    for y in strings:
        if x == y:
            count = count +1
    print(count)        
```

### Array Manipulation

```python
import sys

if __name__ == "__main__":
    n, m = input().strip().split(' ')
    n, m = [int(n), int(m)]
    arr = [0]* (n+1) 
    
    for a0 in range(m):
        a, b, k = input().strip().split(' ')
        a, b, k = [int(a), int(b), int(k)]
        
        arr[a-1] += k
        arr[b] -= k
        
    maxval = 0
    prefixsum=0
    for val in arr:
        prefixsum += val
        if prefixsum>=maxval:
            maxval=prefixsum
    print(maxval)        
```
